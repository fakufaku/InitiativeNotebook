Evolution de l'initiative populaire en Suisse
=============================================

Nous etudions l'evolution du nombre d'initiatives lancees au cours des annees.

Les donnees sont extraites de l'articles wikipedia [Liste des initiatives populaires federales en Suisse](https://fr.wikipedia.org/wiki/Liste_des_initiatives_populaires_f%C3%A9d%C3%A9rales_en_Suisse).

```python
# start pylab and import relevant libraries
%pylab inline
import numpy as np
```

```python
# We need some routine to convert different date formats
from datetime import datetime

def format_date(s):
    months = {'janvier': '01', 'février': '02', 'mars': '03', 'avril': '04', 'mai': '05', 'juin': '06', \
              'juillet': '07', 'août': '08', 'septembre': '09', 'octobre': '10', 'novembre': '11', 'décembre': '12'}
    
    array = s.split()
    
    if len(array) > 1:
        date = array[0]+'-'+str(months[array[1]])+'-'+array[2]
    else:
        date = array[0]
        
    return datetime.strptime(date, '%d-%m-%Y')
```

Extraction et nettoyage des donnees
-----------------------------------

```python
f = open('list.txt')
raw_data = f.readlines()
f.close()
data = [];
for line in raw_data:
    fields = line.split('||')
    
    if not (len(fields) == 8 or len(fields) == 6):
        continue
    
    title = fields[0]
    date1 = format_date(fields[1])
    
    if len(fields) == 8:
        date2 = format_date(fields[5])
        data.append({'titre' : title, 'date_depot' : date1, 'statut' : fields[2].split()[0], 'domaine' : fields[3], \
                     'comite' : fields[4], 'date_votation' : date2, 'oui' : fields[6], 'non' : fields[7]})
    elif len(fields) == 6:
        data.append({'titre' : fields[0], 'date_depot' : date1, 'statut' : fields[2].split()[0], 'domaine' : fields[3], \
                     'comite' : fields[4], 'date_votation' : '', 'oui' : '', 'non' : ''})
```

Traitement des donnees
----------------------

Nous extrayons les initiatives ayant abouti et passe en vote populaire. Nous separons les initiatives acceptees des initiatives rejetees.

```python
year_oui = []
year_non = []
for d in data:
    if d['statut'] == 'O':
        year_oui.append(d['date_votation'].year)
    elif d['statut'] == 'N':
        year_non.append(d['date_votation'].year)
 
year = year_oui + year_non
year_range = range(min(year), max(year),10)
```
